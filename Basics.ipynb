```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map({"Yes":True, "No":False})
```

```python
seens = star_wars.columns[3:9]
```

```python
star_wars[seens].head(5)
```

```python
def seen_to_true(elem):
    if type(elem) != float:
        return True
    else:
        return False
```

```python
star_wars[seens] = star_wars[seens].applymap(seen_to_true)
```

```python
star_wars[seens]
```

```python
for i, movie in enumerate(seens):
    star_wars.rename(columns={movie: f'seen_{i+1}'}, inplace=True)
```

```python
star_wars.iloc[:,9:15].astype(float)

for col in range(9,15):
    star_wars.rename(columns={star_wars.columns[col]:f'ranking_{col-8}'}, inplace=True)
```

```python
star_wars.iloc[0:5,9:15]
```

```python
%matplotlib inline
```

```python
rankings = star_wars.iloc[:,9:15]
means = rankings.agg(np.mean)
means = means.sort_values(ascending=True)

seens = star_wars.iloc[:,3:9]
seens = seens.iloc[:,[int(ind[-1])-1 for ind in means.index]].sum()
```

```python
fig, axs = plt.subplots(1,2,figsize=(10,4))

axs[0].bar(x=[ind[-1] for ind in means.index], height=means)
axs[0].set_title("Movies' Average Rankings")
axs[0].set_xlabel("Movies")
axs[0].set_ylabel("Ranking")

axs[1].bar(x=[ind[-1] for ind in seens.index], height=seens)
axs[1].set_title("Movies' Average Seen")
axs[1].set_xlabel("Movies")
axs[1].set_ylabel("Seen")

plt.show()
```

The number of times movies were seen were reasonable where the best ranked movie had the highest number of watches while the least ranked movie had the lowest.

```python
star_wars_fans = star_wars[star_wars[star_wars.columns[2]]=='Yes']
star_wars_non_fans = star_wars[star_wars[star_wars.columns[2]]!='Yes']
star_wars.shape[0] == star_wars_fans.shape[0] + star_wars_non_fans.shape[0]
```

```python
# Star Wars Fans
rankings = star_wars_fans.iloc[:,9:15]
means = rankings.agg(np.mean)
means = means.sort_values(ascending=True)

seens = star_wars_fans.iloc[:,3:9]
seens = seens.iloc[:,[int(ind[-1])-1 for ind in means.index]].sum()

fig, axs = plt.subplots(2,2,figsize=(10,10))

axs[0,0].bar(x=[ind[-1] for ind in means.index], height=means)
axs[0,0].set_title("Movies' Average Rankings")
axs[0,0].set_ylabel("Fans\nRanking")

axs[0,1].bar(x=[ind[-1] for ind in seens.index], height=seens)
axs[0,1].set_title("Movies' Average Seen")
axs[0,1].set_ylabel("Seen")


# Non Star Wars Fans
rankings = star_wars_non_fans.iloc[:,9:15]
means = rankings.agg(np.mean)
means = means.sort_values(ascending=True)

seens = star_wars_non_fans.iloc[:,3:9]
seens = seens.iloc[:,[int(ind[-1])-1 for ind in means.index]].sum()

axs[1,0].bar(x=[ind[-1] for ind in means.index], height=means)
axs[1,0].set_xlabel("Movies")
axs[1,0].set_ylabel("Non-Fans\nRanking")

axs[1,1].bar(x=[ind[-1] for ind in seens.index], height=seens)
axs[1,1].set_xlabel("Movies")
axs[1,1].set_ylabel("Seen")
plt.show()
```

Regardless of the person being fans or not for the Star Wars series, the majority also loved watching the episode 5 of Star Wars meanwhile the majority also rated episode 3 as the worst among the series.
